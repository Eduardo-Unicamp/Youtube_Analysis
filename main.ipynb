```python
#Importing libraries
import numpy as np
import pandas as pd
from matplotlib import pyplot as plot
import seaborn
import json
```

```python
#Importing the main dataset to be used in the project

youtubebr_df = pd.read_csv(r'data\BR_youtube_trending_data.csv')
youtubebr_df
```

```python
#importing the categories dataframe from json file, normalizing it and adding to a dataframe

with open(r'data\BR_category_id.json') as file:
    data = json.load(file)
categories_df = pd.json_normalize(data, record_path=['items'])
categories_df = categories_df[['id','snippet.title']]
categories_df
```

```python
#changing names for legibility

categories_df.rename(columns={'id':'categoryId','snippet.title':'category'},inplace=True)
categories_df['categoryId'] = categories_df['categoryId'].astype(int)
```

```python

#adding together the two dataframes we're gonna use
merged_youtube_df = pd.merge(youtubebr_df,categories_df,on='categoryId')
merged_youtube_df
```

```python
#remover linhas null
no_null_youtube_df = merged_youtube_df.dropna(how='all')
no_null_youtube_df
```

```python
#checking types for each column
no_null_youtube_df.dtypes
```

```python
#Converting types to more desirable ones

no_null_youtube_df['publishedAt'] = no_null_youtube_df['publishedAt'].str.slice(stop=10)
no_null_youtube_df['publishedAt'] = pd.to_datetime(no_null_youtube_df['publishedAt'],format='%Y-%m-%d')



no_null_youtube_df['trending_date'] = no_null_youtube_df['trending_date'].str.slice(stop=10)
no_null_youtube_df['trending_date'] = pd.to_datetime(no_null_youtube_df['trending_date'],format='%Y-%m-%d')
no_null_youtube_df.head()
```
